```python
q("""
WITH daily AS (
    SELECT 
        date(timestamp) AS day,
        error_code,
        COUNT(*) AS error_count
    FROM device_events
    GROUP BY date(timestamp), error_code
),
with_trailing AS (
    SELECT
        error_code,
        day,
        error_count,
        -- 7-day trailing average, excluding today
        AVG(error_count) OVER (
            PARTITION BY error_code
            ORDER BY day
            ROWS BETWEEN 7 PRECEDING AND 1 PRECEDING
        ) AS trailing_7d_avg
    FROM daily
)
SELECT
    error_code,
    day,
    error_count,
    trailing_7d_avg,
    ROUND(
        CASE 
            WHEN trailing_7d_avg IS NULL OR trailing_7d_avg = 0 
            THEN NULL
            ELSE 1.0 * error_count / trailing_7d_avg
        END,
        2
    ) AS spike_ratio
FROM with_trailing
WHERE trailing_7d_avg IS NOT NULL
  AND error_count >= 1.5 * trailing_7d_avg   -- spike threshold
ORDER BY day, error_code;
""")
```

```python
q("""
WITH pre_post AS (
    SELECT
        fr.firmware_version AS release_version,
        de.error_code,
        CASE
            WHEN date(de.timestamp) BETWEEN date(fr.release_date, '-14 days') 
                                        AND date(fr.release_date, '-1 day')
            THEN 'pre'
            WHEN date(de.timestamp) BETWEEN date(fr.release_date) 
                                        AND date(fr.release_date, '+14 days')
            THEN 'post'
        END AS period,
        COUNT(*) AS error_count
    FROM firmware_releases fr
    JOIN device_events de
      ON date(de.timestamp) BETWEEN date(fr.release_date, '-14 days')
                                AND date(fr.release_date, '+14 days')
    GROUP BY release_version, de.error_code, period
),
pivot AS (
    SELECT
        release_version,
        error_code,
        SUM(CASE WHEN period = 'pre'  THEN error_count ELSE 0 END) AS pre_errors,
        SUM(CASE WHEN period = 'post' THEN error_count ELSE 0 END) AS post_errors
    FROM pre_post
    GROUP BY release_version, error_code
)
SELECT
    release_version,
    error_code,
    pre_errors,
    post_errors,
    (post_errors - pre_errors) AS delta_errors,
    CASE 
        WHEN pre_errors > 0 
        THEN ROUND(1.0 * post_errors / pre_errors, 2)
        ELSE NULL
    END AS post_pre_ratio
FROM pivot
WHERE post_errors >= 50  -- avoid tiny-sample noise
  AND (
        pre_errors = 0
        OR 1.0 * post_errors >= 1.5 * pre_errors     -- 50%+ increase
      )
ORDER BY post_pre_ratio DESC, delta_errors DESC;
""")
```

```python
q("""
WITH events_with_release AS (
    SELECT
        de.firmware_version,
        date(de.timestamp) AS day,
        fr.release_date,
        CAST(
            julianday(date(de.timestamp)) - julianday(date(fr.release_date))
            AS INTEGER
        ) AS days_since_release
    FROM device_events de
    JOIN firmware_releases fr
      ON de.firmware_version = fr.firmware_version
),
daily_by_offset AS (
    SELECT
        firmware_version,
        days_since_release,
        COUNT(*) AS error_count
    FROM events_with_release
    WHERE days_since_release >= 0   -- only post-release
    GROUP BY firmware_version, days_since_release
)
SELECT
    firmware_version,
    days_since_release,
    error_count
FROM daily_by_offset
ORDER BY firmware_version, days_since_release;
""")
```

```python
rma_by_error_df = q("""
SELECT
    error_code,
    COUNT(*) AS ticket_count,
    SUM(rma_issued) AS rma_count,
    ROUND(1.0 * SUM(rma_issued) / COUNT(*), 3) AS rma_rate
FROM support_tickets
GROUP BY error_code
ORDER BY rma_rate DESC;
""")
```

```python
rma_by_firmware_df = q("""
WITH ticket_events AS (
    SELECT
        t.ticket_id,
        t.device_id,
        t.error_code,
        t.rma_issued,
        t.created_at,
        e.firmware_version
    FROM support_tickets t
    JOIN device_events e ON t.device_id = e.device_id AND t.error_code = e.error_code
       AND e.timestamp <= t.created_at
)
SELECT
    firmware_version,
    COUNT(*) AS tickets,
    SUM(rma_issued) AS rmas,
    ROUND(1.0 * SUM(rma_issued) / COUNT(*), 3) AS rma_rate
FROM ticket_events
GROUP BY firmware_version
ORDER BY rma_rate DESC;
""")
```

```python
error_rates_df = q("""
WITH device_base AS (
    SELECT COUNT(DISTINCT device_id) AS total_devices
    FROM device_events
),
errors_by_fw AS (
    SELECT firmware_version, COUNT(*) AS error_events
    FROM device_events
    GROUP BY firmware_version
)
SELECT
    e.firmware_version,
    e.error_events,
    ROUND(1.0 * e.error_events / d.total_devices * 10000, 2) AS error_rate_per_10k
FROM errors_by_fw e
CROSS JOIN device_base d
ORDER BY error_rate_per_10k DESC;
""")
```

```python
spikes_df = q("""
WITH daily AS (
    SELECT 
        error_code,
        DATE(timestamp) AS day,
        COUNT(*) AS error_count
    FROM device_events
    GROUP BY 1,2
),
rolled AS (
    SELECT
        error_code,
        day,
        error_count,
        AVG(error_count) OVER (
            PARTITION BY error_code
            ORDER BY day
            ROWS BETWEEN 7 PRECEDING AND 1 PRECEDING
        ) AS trailing_7d_avg
    FROM daily
)
SELECT
    error_code,
    day,
    error_count,
    trailing_7d_avg,
    ROUND(CASE 
        WHEN trailing_7d_avg IS NULL THEN NULL
        ELSE 1.0 * error_count / trailing_7d_avg
    END, 2) AS spike_ratio
FROM rolled
WHERE trailing_7d_avg IS NOT NULL
  AND spike_ratio >= 1.5
ORDER BY spike_ratio DESC;
""")
```

```python
stability_df = q("""
SELECT
    e.firmware_version,
    CAST(julianday(e.timestamp) - julianday(r.release_date) AS INT) AS days_since_release,
    COUNT(*) AS error_count
FROM device_events e
JOIN firmware_releases r
  ON e.firmware_version = r.firmware_version
GROUP BY 1,2
ORDER BY 1,2;
""")
```

```python
pre_post_df = q("""
WITH pre_post AS (
    SELECT
        fr.firmware_version AS release_version,
        de.error_code,
        CASE
            WHEN date(de.timestamp) BETWEEN date(fr.release_date, '-14 days')
                                        AND date(fr.release_date, '-1 day')
                THEN 'pre'
            WHEN date(de.timestamp) BETWEEN date(fr.release_date)
                                        AND date(fr.release_date, '+14 days')
                THEN 'post'
        END AS period
    FROM firmware_releases fr
    JOIN device_events de
      ON date(de.timestamp) BETWEEN date(fr.release_date, '-14 days')
                                AND date(fr.release_date, '+14 days')
),
agg AS (
    SELECT
        release_version,
        error_code,
        SUM(CASE WHEN period = 'pre'  THEN 1 ELSE 0 END) AS pre_errors,
        SUM(CASE WHEN period = 'post' THEN 1 ELSE 0 END) AS post_errors
    FROM pre_post
    GROUP BY release_version, error_code
)
SELECT
    release_version,
    error_code,
    pre_errors,
    post_errors,
    (post_errors - pre_errors) AS delta_errors,
    CASE 
        WHEN pre_errors > 0 
        THEN ROUND(1.0 * post_errors / pre_errors, 2)
        ELSE NULL
    END AS post_pre_ratio
FROM agg
WHERE post_errors >= 50        -- keep only meaningful post volume
ORDER BY release_version, error_code;
""")
```
